```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1

```python
df = pd.read_excel('motorCC.xlsx')
x = np.array(df.iloc[:,0])
x = x.reshape(x.shape[0],1)
y = np.array(df.iloc[:,1])
y = y.reshape(y.shape[0],1)
reg = LinearRegression().fit(x[8:], y[8:])
a = reg.coef_
b = reg.intercept_

plt.figure(figsize=(10,8))
sns.scatterplot(x[8:,0],y[8:,0],label='experimental');
t = np.arange(0,20,0.01)
r = a*t+b
sns.lineplot(t,r[0,:], color='r',label='teórico');
plt.legend()
plt.savefig('linear_va_vt.png')
```
